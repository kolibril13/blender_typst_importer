```python
from typst_importer.typst_to_svg import typst_express
from typst_importer.notebook_utils import display_svg

def eq(equation: str, name: str, color: str, position=(0, 0, 0)):

    typst_code = f"""
    #set text({color})
    $ {equation} $
    """
    collection = typst_express(
        typst_code,
        origin_to_char=True,
        convert_to_mesh=True,
        name=name,
        position=position
    )
    display_svg(collection.processed_svg, width="230px")
    return collection
eq("a", "Start", "aqua", (0, 0, 0.1))
eq("a = b + c", "End", "olive", (0, -0.8, 0));

eq("a = b + c ", "Start", "aqua", (0, -2, 0.1))
eq("a -b = c", "End", "olive", (0, -2.8, 0));

eq("a -b = c", "Start", "aqua", (0, -4, 0.1))
eq("a - 3 r_(beta) = c", "End", "olive", (0, -4.8, 0));


eq("a - 3 r_(beta) = c", "End", "aqua", (0, -6, 0));
eq("(a - 3 r_(beta))/c=1", "End", "olive", (0, -7.8, 0));

```
